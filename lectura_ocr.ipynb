## Posibles variaciones del código base:

Idioma: ocrmypdf -l spa documento_entrada.pdf documento_salida.pdf

Multiples idiomas: ocrmypdf -l spa+eng documento_entrada.pdf documento_salida.pdf

Imagenes: ocrmypdf --image-dpi 300 documento_entrada.pdf documento_salida.pdf

Inclinación: ocrmypdf --deskew documento_entrada.pdf documento_salida.pdf

Limpieza: ocrmypdf --clean documento_entrada.pdf documento_salida.pdf

Optimizar: ocrmypdf --optimize 3 documento_entrada.pdf documento_salida.pdf
